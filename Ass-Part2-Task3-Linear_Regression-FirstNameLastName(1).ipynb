# Import Libraries  

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import polyfit
from scipy.stats import linregress
```

# Import Data

```python

# Load the dataset with explicit encoding
data = pd.read_csv('bottle.csv', encoding='latin1', low_memory=False)
```

# Data management 

```python
# Select relevant columns
selected_data = data[['Salnty', 'T_degC']].dropna()  # Remove rows with missing values
```

# Linear Regression

```python

# Scatter plot with regression line
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Salnty', y='T_degC', data=selected_data, alpha=0.5)
sns.regplot(x='Salnty', y='T_degC', data=selected_data, scatter=False, color='red', line_kws={'label':'Regression Line'})
plt.title('Scatter Plot with Regression Line')
plt.xlabel('Salinity (Salnty)')
plt.ylabel('Temperature (T_degC)')
plt.legend()
plt.show()
```

```python
#  Regression Analysis Results

# Linear regression analysis
slope, intercept, r_value, p_value, std_err = linregress(selected_data['Salnty'], selected_data['T_degC'])

# Print results
print("Regression Analysis Results:")
print(f"Slope (m): {slope}")
print(f"Intercept (b): {intercept}")
print(f"Correlation Coefficient (r): {r_value}")
print(f"P-value: {p_value}")
print(f"Standard Error: {std_err}")
```

```python
# RESIDUAL PLOT
# Calculate residuals
residuals = selected_data['T_degC'] - (intercept + slope * selected_data['Salnty'])

# Residual plot
plt.figure(figsize=(12, 8))
sns.scatterplot(x=selected_data['Salnty'], y=residuals, alpha=0.5)
plt.axhline(y=0, color='red', linestyle='--', label='Residuals Mean Line')
plt.title('Residual Plot')
plt.xlabel('Salinity (Salnty)')
plt.ylabel('Residuals')
plt.legend()
plt.show()
```
