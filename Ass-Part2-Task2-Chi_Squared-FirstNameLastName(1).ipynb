# Import Libraries  

```python
import pandas as pd
from scipy.stats import chi2_contingency
import seaborn as sns
import matplotlib.pyplot as plt
```

# Import Data

```python

# Load the dataset with explicit encoding
data = pd.read_csv('bottle.csv', encoding='latin1', low_memory=False)
```

# Data management 

```python
# Select the relevant columns
selected_data = data[['T_qual', 'S_qual']].dropna()  # Remove rows with missing values

# Create a cross-tabulation table
cross_tab = pd.crosstab(selected_data['T_qual'], selected_data['S_qual'])

# Perform Chi-Squared Test
chi2, p_value, _, expected = chi2_contingency(cross_tab)

# Print the results
print("Cross Tab (observed numbers):")
print(cross_tab)
print("\nCross Tab (Percentages):")
print(cross_tab.div(cross_tab.sum(axis=1), axis=0) * 100)  # Calculate percentages
print("\nCross Tab (expected numbers):")
print(pd.DataFrame(expected, index=cross_tab.index, columns=cross_tab.columns))
```

# Chart/Plot

```python


# Plotting a heatmap of the cross-tabulation table
plt.figure(figsize=(12, 8))
sns.heatmap(cross_tab, annot=True, fmt='d', cmap='coolwarm', cbar=True)
plt.title('Heatmap of Cross Tabulation (Observed Numbers)')
plt.xlabel('Salinity Quality (S_qual)')
plt.ylabel('Temperature Quality (T_qual)')

plt.draw()

plt.savefig('plot-chart.png')

```
